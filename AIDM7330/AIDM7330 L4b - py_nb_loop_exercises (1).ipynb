## Exercise 1
Rewrite the program using:
- only one input 

````python
import random
extractedNumber = int(random.randint(1,10))
myGuess = None

while myGuess != extractedNumber:
    myGuess = int(input('Your guessed number:'))
    if myGuess > extractedNumber:
        print('Guess a smaller number')
    if myGuess < extractedNumber:
        print('Guess a bigger number')
    if myGuess == extractedNumber:
        print('Bravo!')
        break
print(extractedNumber)
````

## Exercise 2
Rewrite the program using:
- only one input 
- continue statement

````python
import random
extractedNumber = int(random.randint(1,10))
myGuess = None
while myGuess != extractedNumber:
    myGuess = int(input('Your guessed number:'))
    if myGuess > extractedNumber:
        print('Guess a smaller number')
        continue
    if myGuess < extractedNumber:
        print('Guess a bigger number')
        continue
    if myGuess == extractedNumber:
        print('Bravo!')
        break #continue
print(extractedNumber)
````

## Exercise 3
Rewrite the program using:
- only one input 
- continue statement
- boolean variable ```guessedCorrectly``` to control the loop and check if the user guessed the right number


````python
import random
extractedNumber = int(random.randint(1,10))
guessedCorrectly = False

while not(guessedCorrectly == True): # Shortcut not(guessedCorrectly)
    myGuess = int(input('Your guessed number:'))
    if myGuess > extractedNumber:
        print('Guess a smaller number')
        continue
    if myGuess < extractedNumber:
        print('Guess a bigger number')
        continue
    if myGuess == extractedNumber:
        print('Bravo!')
        guessedCorrectly = True
        continue # Is it needed?
print(extractedNumber)
````

## Exercise 4
Find the minimum element in a list

````python
# Solution with a semantic error
minElement = 0
for number in [8, 56, 75, 9, 10]:
    if number <= minElement:
        minElement = number
print('Minimum element value:', minElement)
````

````python
# A correct solution
minElement = None
for number in [8, 56, 75, 9, 10]:
    if minElement is None: # Similar to == but more powerful
        minElement = number
    elif number <= minElement:
        minElement = number
print('Minimum element value:', minElement)
````

We can rewrite the maximum function to work also with negative numbers

````python
# A correct solution
maxElement = None
for number in [-8, -56, -75, -9, -10]:
    if maxElement is None:
        maxElement = number
    elif number >= maxElement:
        maxElement = number
print('Maximum element value:', maxElement)
````
