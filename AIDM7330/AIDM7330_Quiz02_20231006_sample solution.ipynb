## Exercise n. 1 (20 Marks)
Request 2 numbers to the user. Define a function that print the minimum between the two numbers passed as parameters.
Try to use `if … elif … else` statements and avoid using `min()` function.

```python
def minNum(a,b):
    if a < b:
        print(a)
    elif a > b:
        print(b)
    else:
        print('a equals b')

firstNumber = int(input('Please input the first number:'))
secondNumber = int(input('Please input the second number:'))
minNum(firstNumber,secondNumber)
```

## Exercise n.2 (20 Marks)
Request a single character to the user. Define a function that receives the character as parameter. The function will print if the character is a vowel (e.g. a, e, i, o, u) or not.

```python
def vowel(char):
    allVowels = ['a','e','i','o','u']
    if char.lower() in allVowels:
        print('This character is a vowel.')
    else:
        print('This character is not a vowel')
        
character = input('Please input the character you want to check:')
vowel(character)
```

## Exercise n. 3 (20 Marks)
Given a list of numbers, define a function that computes (and returns) the average of all the numbers in the list.
Use the `for` statement. Appropriate conditional checks are needed.

```python
def num_avg(alist):
    total = 0
    list_len = len(alist)
    if list_len > 0: # conditional check #1
        try:         # conditional check #2
            for i in alist:
                total += float(i)
            avg = total / list_len
            return avg
        except ValueError:
            print("has characters")
    else:
        print("list is of length 0")
```

```python
test_list1 = [1, 2, 3]
result1 = num_avg(test_list1)
result1
```

```python
test_list2 = []
result2 = num_avg(test_list2)
result2
```

```python
test_list3 = [1, 2, "a"]
result3 = num_avg(test_list3)
result3
```

## Exercise n. 4 (20 Marks)
Textual histogram generator. Define a function that, given a list of numbers, prints an histogram based on those numbers. Use the symbol `>` to draw.
Example:
Given the list [2,7,8,4] it should produce the following histogram:

`>>`

`>>>>>>>`

`>>>>>>>>`

`>>>>`

```python
def histogram(numberList):
    for number in numberList:
        print('>' * int(number)) #easy solution multiplication of strings
        #for j in range(number): #alternative to *
        #    print('>')

list1 = [2,7,8,4]
histogram(list1)
```

## Exercise n. 5 (20 Marks)
Same as exercise n. 4 but the function should decide how to order the list give a parameter. The parameter should allow values “Ascending” or “Descending”. It should be used to produce an appropriate histogram based on the ordered list.
Example:
Given the list [2,6,8,4] and ordering “Ascending” it should produce the following histogram:

`>>`

`>>>>`

`>>>>>>`

`>>>>>>>>`


```python
def histogram(numberList, ascending): #two parameters
    if ascending == True:
        sortedList = sorted(numberList, reverse=False)
    else:
        sortedList = sorted(numberList, reverse=True)
        
    for number in sortedList:
        print('>' * int(number))

list1 = [2,6,8,4]
orderAscending = True
histogram(list1, orderAscending)
```
