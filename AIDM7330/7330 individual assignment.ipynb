```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
from time import sleep
from time import time
from random import randint
```

```python
#find the market cap
headers = {"Accept-Language": "en-US, en;q=0.5"}
url = 'https://crypto.com/price?page=24'
page = requests.get(url, headers = headers)
print(page.status_code)

html_soup = BeautifulSoup(page.content, 'html.parser')
cryptocurrency_containers = html_soup.find_all('tr', class_ = 'css-1cxc880')

# initialize
market_cap_list = []

domain = 'https://crypto.com'
for container in cryptocurrency_containers:
    sub_url = container.select('.css-1sem0fc a')[0]['href']
    # url of market cap page
    cryptocurrency_url = domain + sub_url
    marketCap_html = requests.get(cryptocurrency_url, headers = headers)
    # Pause the loop
    sleep(randint(3,6))

    # extract market cap information
    cryptocurrency_soup = BeautifulSoup(marketCap_html.content, 'html.parser') # parser of individual page
    info_li = cryptocurrency_soup.find_all('div', class_ = 'css-7zicxd')
    if len(info_li) > 0:
      market_cap = info_li[0].find('p', class_ = 'chakra-text css-1c8c51m').get_text()
      market_cap_list.append(market_cap)

    else:
      print('No reviews found, please check')


print(market_cap_list)



```

```python
marketCap_df = pd.DataFrame({'Market Cap': market_cap_list})
```

```python
# store the scraped information in a dataframe format
result = pd.concat([test_df, marketCap_df], axis=1)
result
```
