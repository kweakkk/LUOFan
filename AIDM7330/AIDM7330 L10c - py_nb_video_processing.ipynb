## Concatenating clips

A simple way of putting clips together is to concatenate them, to play them one after the other in a single long clip.

```python
# Load summer.mp4 and select the subclip 00:00:50 - 00:00:60
clipSummer = VideoFileClip(dataPath + 'summer.mp4')
clipOcean = VideoFileClip(dataPath + 'ocean.mp4')

clipOceanDuration = clipOcean.duration
subClipOcean = clipOcean.subclip(0, clipOceanDuration / 4)

subClipSummer = clipSummer.subclip(0,10)

finalClip = concatenate_videoclips([subClipOcean,subClipSummer])

finalClip.write_videofile(dataPath + 'summer_edited.mp4')

ipython_display(dataPath + 'summer_edited.mp4', width=280)
```

## Adding Transitions

To add transitions between clips, set a `length` of transition and use it as a parameter for one of the following methods:
- `crossfadein()`
- `crossfadeout()` (very slow)

Orher transitions are planned by the authors but still not implemented. Programmers can create their own tranistions using `fx` effects.

```python
length = 1 #duration of fade in seconds

finalClip = concatenate([subClipOcean,
                         subClipSummer.crossfadein(length)],#crossfadein :going to appear
                         padding=-length, method="compose")
finalClip.write_videofile(dataPath + 'summer_edited_transition.mp4')

ipython_display(dataPath + 'summer_edited_transition.mp4', width=280)
```

## Stacking clips

Another way of compositing clips is to stack them, to play them side by side in a single larger clip.

```python
clipSummer = VideoFileClip(dataPath + 'summer.mp4').subclip(0,5)
clip1 = clipSummer.margin(10) # add 10px contour
clip2 = clip1.fx(vfx.mirror_x)
clip3 = clip1.fx(vfx.mirror_y)
clip4 = clip1.resize(0.60) # downsize 60%

final_clip = clips_array([[clip1, clip2],
                          [clip3, clip4]])

final_clip.resize(width=480).write_videofile(dataPath + 'my_stack.mp4')

ipython_display(dataPath + 'my_stack.mp4', width=280)
```
