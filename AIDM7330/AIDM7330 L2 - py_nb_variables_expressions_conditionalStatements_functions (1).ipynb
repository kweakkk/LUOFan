## Alternative executions with else
Be careful, semantic errors hide everywhere... also in simple code.
Where is in the following?

```python
x = 10
if x < 10:
    print('Smaller') # x < 10 -> True
else:
  if x==10:
    print('same')
  else:
    print('Bigger') # x < 10 -> not(True) equal or greater
print('Finish')
#not all the clause will be excuated
#this code can check 3 values
```

A more complex example, using Boolean values

```python
rate = input('Input the rate:')
hours = input('Input hours:')
stop = False

try:
    numRate = float(rate)
except:
    print('Rate is not a number')
    stop = True # Flag

try:
    numHours = float(hours)
except:
    print('Hours is not a number')
    stop = True

if not(stop == True):# which indicated that stop== False
    if numHours <= 40:
        salary = numRate * numHours
    else:
        baseSalary = numRate * 40
        otHours = numHours - 40
        otRate = numRate * 1.5
        otSalary = otRate * otHours
        #print('otSalary:', otSalary)
        salary = baseSalary + otSalary

        # stop is a mark help us manage the error
        #we can not use continue as the varible name, because it is a name of python use

    print(salary)
```

# Functions
How to abstract and reuse the code

hide the complexity of code

```python
# Function definition
def happyBirthday(): #定义一个function 执行无结果 this function have not parameter,a variable 括号里没变量
    print('Happy Birthday to you!')
    print('Happy Birthday to you!')
    print('Happy Birthday, dear Emily')
    print('Happy Birthday to you!')
```

```python
# Function call
happyBirthday()
```

with parameters

```python
def happyBirthday(person):
    print('Happy Birthday to you!')
    print('Happy Birthday to you!')
    print('Happy Birthday, dear', person)
    print('Happy Birthday to you!')
```

```python
happyBirthday('Joey')
```

multiple parameters

```python
def happyBirthday(person, years):
    print('Happy Birthday to you!')
    print('Happy Birthday to you!')
    print('Happy Birthday, dear', person)
    print('Happy', str(years), 'Birthday to you!')
```

```python
personName = 'Joey'
age = '44'
happyBirthday(personName, age)
```

## Variable scope
Or where the variable is available for computation

```python
def testScope(parameterVar):
    resultVar = parameterVar + 10
    return resultVar
```

```python
print(resultVar)
```

```python
testScope(10)
```

```python
newVar = 10
returnedVar = testScope(newVar)
print(returnedVar)
```

```python
resultVar = 10
returnedVar = testScope(resultVar)
print(returnedVar, resultVar)
```
