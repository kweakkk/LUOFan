### Exercise n. 1 (20 Marks)
Define a function that takes two sets as parameters, then concatenate these two sets to create a new dictionary as returned value. Use the `for` loop.

```python
def joinSets(set1,set2):
    resultSet = set()
    for element in set1:
      resultSet.add(element)
    for element in set2:
      resultSet.add(element)
    return resultSet

a = {'money', 'candy'}
b = {'tissue', 'smartphone'}
result = joinSets(a,b)
print(result)
```

### Exercise n.2 (20 Mark)
Define a function that, given a string as parameter, returns a dictionary containing the frequency of the characters in the string.

```python
def charFreq(str):
    charMap = {}
    for character in str:
        if character in charMap:
            charMap[character] += 1
        else:
            charMap[character] = 1
    return charMap

str = 'xyxyzz'
charFreq(str)
```

### Exercise n. 3 (20 Mark)
Define a function that, given a string as parameter, returns a dictionary containing the frequency of the words in the string.

```python
def wordFreq(text):
    wordMap = {}
    for word in text.split(' '):
        if word in wordMap:
            wordMap[word] += 1
        else:
            wordMap[word] = 1
    return wordMap

text = 'the meal in the canteen was very very nice'
wordFreq(text)
```
